##### BAR PLOT TO SHOW RELATION OF TARGET WITH OTHER VARIABLES

```python
#Get Correlation of "Target" with other variables:

colors = ['#8000ff','#ff3d03','#1414ff']
plt.figure(figsize=(15,8))
df.corr()['target'].sort_values(ascending = False).plot(kind='bar',color=colors)
```

```python
# Create Train & Test Data

from sklearn.model_selection import train_test_split
x=df[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']]
y=df[["target"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

##### K NEIGHBORS CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
clf=KNeighborsClassifier(n_neighbors=3)
clf.fit(x,y)
y_pred=clf.predict(x_test)
score=accuracy_score(y_test,y_pred)
print(" Accuracy with KNeighborsClassifier is",score)
```
