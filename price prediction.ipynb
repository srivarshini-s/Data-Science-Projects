```python
cust_item_matrix = df.pivot_table(index='CustomerID', columns='StockCode', values='Quantity',aggfunc='sum')
cust_item_matrix.head()
```

```python
cust_item_matrix = cust_item_matrix.applymap(lambda x: 1 if x > 0 else 0)
cust_item_matrix.head()
```

```python
cust_item_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_sim_matrix = pd.DataFrame(cosine_similarity(cust_item_matrix))
user_sim_matrix
```

```python
user_sim_matrix.shape
```

```python
user_sim_matrix.columns = cust_item_matrix.index

user_sim_matrix['CustomerID'] = cust_item_matrix.index

user_sim_matrix = user_sim_matrix.set_index('CustomerID')
user_sim_matrix.head()
```

```python
def get_items_to_recommend_cust(cust_a): 
  most_similar_user = user_sim_matrix.loc[cust_a].sort_values(ascending=False).reset_index().iloc[1, 0]
  items_bought_by_cust_a = set(cust_item_matrix.loc[cust_a].iloc[cust_item_matrix.loc[cust_a].to_numpy().nonzero()].index)
  items_bought_by_cust_b = set(cust_item_matrix.loc[most_similar_user].iloc[cust_item_matrix.loc[most_similar_user].to_numpy().nonzero()].index)
  items_to_recommend_to_a = items_bought_by_cust_b - items_bought_by_cust_a
  items_description = df.loc[df['StockCode'].isin(items_to_recommend_to_a), ['StockCode', 'Description']].drop_duplicates().set_index('StockCode')
  return items_description
```

```python
def get_top_similar_items(item):
  top_10_similar_items = list(item_sim_matrix.loc[item].sort_values(ascending=False).iloc[:10].index)
  top_10 = df.loc[df['StockCode'].isin(top_10_similar_items), ['StockCode', 'Description']].drop_duplicates().set_index('StockCode').loc[top_10_similar_items]
  return top_10
```
